## Subgraphs

```python
er = nx.erdos_renyi_graph(100, 0.15)
nx.draw(er, with_labels=True)
```

```python
sub = nx.subgraph(er, [54, 35, 45, 76, 59, 90, 4, 99])
nx.draw(sub, node_color='c', node_size=800, with_labels=True)
```

## Find the number of triangles in a graph

```python
G = nx.complete_graph(5)
nx.draw(G, with_labels=True, node_color='c')
```

```python
# print number of triangles at node 0
print(nx.triangles(G,0))
```

```python
# print the number of triangles in the graph
print(nx.triangles(G))
```

```python
# print the list of triangles in the graph
print(list(nx.triangles(G,(0,1)).values()))
```

## Facebook Analysis

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
import os
os.listdir()
```

```python
# read in file
fb = nx.read_edgelist('facebook_combined.txt', create_using=nx.Graph(), nodetype=int)
```

```python
print(nx.info(fb))
```

```python
# create a spring-layout of fb
pos = nx.spring_layout(fb)

plt.figure(figsize=(10,7))

plt.axis('off')

nx.draw_networkx(fb, pos=pos, with_labels=False, node_size=35)
```

```python
print(nx.betweenness_centrality(fb, normalized=True))
```

```python
print(fb.degree()) # find the number of degree
```

```python
print(fb.neighbors(107))
```

```python
# find the order and size of our network
print(fb.order())
print(fb.size())
```

```python
# find the number of triangles in the network
print(nx.triangles(fb))
```

```python
print(nx.clustering(fb))
```

```python
# draw histogram of degrees
plt.figure(figsize=(10,7))
hst = nx.degree_histogram(fb)
plt.hist(hst, bins=10, color='c')
```

```python
# draw a bar chart of the betweenness centrality
bc = nx.betweenness_centrality(fb)
plt.bar(range(len(bc)), bc.values(), align='center')
```
