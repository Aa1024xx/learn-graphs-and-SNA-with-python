## Defining Functions

```python
# define a function in python
def draw_graph():
    G = nx.Graph()
    G.add_edges_from([(1,2),(2,3),(1,3),(1,4)])
    nx.draw(G, node_size=800, node_color='cyan', with_labels=True)
```

```python
# call the draw_graph function
draw_graph()
```

```python
# define a new function called draw_barabasi_albert
def draw_barabasi_albert():
    G = nx.barabasi_albert_graph(100,10)
    hst = nx.degree_histogram(G)
    plt.hist(hst, bins=20, color='red')
```

```python
draw_barabasi_albert()
```

```python
# creat nodes using a custom function
# define our function called simple-graph

def simple_graph():
    # create an empty graph
    G = nx.Graph()
    
    # add edges
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    # draw the graph
    nx.draw(G, with_labels=True, node_color='y', node_size=800)
```

```python
# call simple_graph function
simple_graph()
```

```python
# delete nodes using a custion function
def delete_nodes_in_graph():
    G = nx.Graph()
    
    # add 3 edges to the graph
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G, with_labels=True, node_color='y', node_size=800)
    
    # remove node B from the graph
    G.remove_node('B')
    
    nx.draw(G, with_labels=True, node_color='c', node_size=800)
```

```python
delete_nodes_in_graph()
```

```python
# delete edges using a custom function
def delete_specific_edge_in_graph():
    G = nx.Graph()
    
    # add 3 edges to the graph
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G, with_labels=True)
    plt.show()
    
    #remove 2 edges from the graph
    G.remove_edge('A','B')
    G.remove_edge('A','C')
    
    nx.draw(G, with_labels=True)
    plt.show()
```

```python
delete_specific_edge_in_graph()
```

```python
# custom node size and node color using a custom function
def custom_node_size_and_node_color():
    G = nx.Graph()
    
    # add 3 edges to the graph
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G)
    plt.show()
    
    # define the nodelist
    nodelst = ['A','B','C']
    
    # define the colorlist
    colorlst = ['c','m','y']
    
    # define the nodesizelist
    nodesize = [1000, 2000, 3000]
    
    nx.draw(G, with_labels=True, node_color=colorlst, node_size=nodesize, nodelist=nodelst)
```

```python
custom_node_size_and_node_color()
```

```python
# custom edge colors using a custom function
def custom_edge_colors():
    G = nx.Graph()
    
     
    # add 3 edges to the graph
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G)
    plt.show()
    
    # create list of edge colors
    edgecol = ['c','y','m']
    
    # create node sizes
    nodesize = [1000, 2000, 3000]
    
    # create the edgelist
    edgelst = [('A','B'),('B','C'),('C','A')]
    
    nx.draw(G, edgelist=edgelst, edge_color=edgecol, with_labels=True)
    plt.show()
```

```python
custom_edge_colors()
```

```python
hartford = nx.read_edgelist('edge.txt', create_using=nx.DiGraph(), nodetype=int)

N, E = hartford.order(), hartford.size()

avg_deg = float(E)/N

print('Nodes:', N)
print('Edge:', E)

print('Average Degree:', avg_deg)
```
